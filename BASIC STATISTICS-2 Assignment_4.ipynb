Problem Statement : Hospital Patient Data Analysis  

Context:                                   

A hospital maintains patient records including admission details, department, diagnosis, doctor, and bill amount. You have two datasets: one with patient info and another with billing details. Some patients have blank bill amounts, and there are multiple rows for the same patient due to follow-ups.


Tasks:

1.	Load the patient dataset and show summary with info().

```python
import pandas as pd
```

```python
Biliing_data = pd.read_csv("Billing_Data.csv")
Patient_data = pd.read_csv("Patient_Data.csv")
```

```python
print("Billing Data")
Biliing_data.info()
Biliing_data
```

```python
print("Patient Data")
Patient_data.info()
Patient_data
```

2.	Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].

```python
Patient_data.loc[:,["PatientID","Department","Doctor","BillAmount"]]
```

3.	Drop administrative columns like ['ReceptionistID', 'CheckInTime'].

```python
Patient_data.drop(["ReceptionistID","CheckInTime"], axis=1,inplace=True)
```

```python
Patient_data
```

4.	Use groupby to find total bill amount per department.

```python
total_bill_per_dept = Patient_data.groupby('Department')['BillAmount'].sum().reset_index()
print("Total Biil Amount per Department")
print(total_bill_per_dept)

```

5.	Remove duplicate patient records based on PatientID.

```python
Patient_data.drop_duplicates(subset='PatientID', keep='first', inplace=True)
Patient_data
```

6.	Fill missing BillAmount values with the mean bill amount.

```python
mean_bill = Patient_data["BillAmount"].mean()
```

```python
Patient_data["BillAmount"] = Patient_data["BillAmount"].fillna(mean_bill) 
```

```python
Patient_data
```

7.	Merge the billing dataset with patient dataset on PatientID.

```python
merged_bill = pd.merge(Biliing_data,Patient_data , on='PatientID', how='left')
merged_bill
```

8.	Concatenate an additional DataFrame that contains new patients for the current week (row-wise).

```python
New_patients_data = {'PatientID': [106, 107],
                      'Name': ['Frank', 'Grace'],
                      'Department': ['Oncology', 'Radiology'],
                      'Doctor': ['Dr. Kane', 'Dr. Turner'],
                      'BillAmount': [8000.0, 9056.0]}
new_patients_df = pd.DataFrame(New_patients_data)
```

```python
merged_bill = pd.concat([merged_bill, new_patients_df], ignore_index=True)
```

```python
merged_bill
```

9.	Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).

```python
insurance_covered = [True, False, True, False, True, True,False]  
final_amount = [4500.0, 1000.0, 6750.0, 5580.0, 8564.0, 4500.0, 1306.0]  
```

```python
merged_bill['InsuranceCovered'] = insurance_covered
merged_bill['FinalAmount'] = final_amount

print(merged_bill)
```
